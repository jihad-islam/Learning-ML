```python
# linspace method ta amra use korte pari, jodi amra kono range er moddhe specific number of equally spaced number generate korte chai.

# linspace er syntax holo: np.linspace(start, stop, num)

np.linspace(10,20,5)
```

```python
# eye() method diye amra identity matrix generate korte pari. identity matrix holo square matrix, jekhane diagonal er element gulo 1 and baki sob element gulo 0 thake.

'''
I3 = [[1. 0. 0.]
      [0. 1. 0.]
      [0. 0. 1.]]
'''

np.eye(3) # 3x3 identity matrix generate korbe
```

## random module or method (vvi)

```python
# random() method diye amra random number generate korte pari.

# rand() method ta uniform distribution er random number generate kore. 

np.random.rand() # ekhane random ekta value pabo eta 0 to 1 er moddhe hobe.
```

```python
np.random.rand(5) # ekhane 5 ta random value pabo, 0 to 1 er moddhe hobe.
```

```python
np.random.rand(10,2) # ekhane 10 row and 2 column er random value pabo, 0 to 1 er moddhe hobe.
```

```python
# negative chaile randn hobe. 

# randn() method ta normal distribution or gaussian distribution er random number generate kore.

np.random.randn() # ekhane random ekta value pabo eta -1 to 1 er moddhe hobe.
```

### randint

```python
# ekta range er moddhe random number generate korte hole amra randint method use korte pari.

# randint er syntax holo: np.random.randint(low, high=None, size=None, dtype=int)

np.random.randint(1,10) # ekhane 1 to 10 er moddhe random ekta value pabo, 10 ta include hobe na.
```

```python
np.random.randint(1,10,5) # ekhane 1 to 10 er moddhe random 5 ta value pabo, 10 ta include hobe na.
```

```python
np.random.randint(1,10,[5,2]) # ekhane 1 to 10 er moddhe random 5x2 er matrix pabo, 10 ta include hobe na.
```

## shape manipulation

```python
# amra jokhon real project er dataset niye kaj korbo tokhon dataset er datagula kon dimension(row * column) e ase sheta directly dekhte parbo na. dimension dekhte shape attribute use korte hobe.

random_array = np.random.randint(10,21,[6,3,4]) # ekhane 10 to 21 er moddhe random 6x3x4 er 3d matrix pabo, 21 ta include hobe na.
random_array.shape 
```

```python
random_array 
```

```python
# reshape method diye amra kono array er shape change korte pari. but total number of element same thakbe. for example amader random_array er shape 6x3x4, mane total element 6*3*4=72. amra chaile 72 ta element ke 2d array te convert korte pari, for example 8x9 or 12x6 etc. but total element same thakbe.

# most important shape change korte hole total element number same thakte hobe.

new_shape_array = random_array.reshape(12,6) 

new_shape_array
```

```python
# find out the maximum and minimum value from the new_shape_array.

new_shape_array.max() # maximum value
```

```python
new_shape_array.min() # minimum value
```

```python
# maximum and minimum value er index find out korte hole amra argmax() and argmin() method use korte pari. 

# but argmax() and argmin() method ta 1D array er jonno kaj kore. tai amra new_shape_array ke 1D array te convert kore nibo.

new_shape_array_1d = new_shape_array.reshape(72) # ekhane 72 ta element ke 1D array te convert korlam.

new_shape_array_1d
```

```python
new_shape_array_1d.argmax() # maximum value er index
```

```python
new_shape_array_1d.argmin() # minimum value er index
```
