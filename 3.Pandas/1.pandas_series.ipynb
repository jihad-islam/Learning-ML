# Pandas Series

```python
import numpy as np
import pandas as pd
```

```python
#pandas er theory mainly 2 ta: pandas series and pandas dataframe.
#panda series mainly 1d array niye kaj kore and pandas dataframe nd array niye kaj kore.

data = np.arange(101,106)
data
```

```python
# ei numpy array ke pandas series e convert korbo.
# ekhane Series holo ekta class. and class e parameter pass kora jay.

pd_series = pd.Series(data = data)
pd_series
```

```python
# amra panda series e index and value both dekhte pari.

# index dekhte .index nam e ekta attribute use korte hoy.
pd_series.index
```

```python
# value dekhte .values nam e ekta attribute use korte hoy.
pd_series.values
```

```python
# amra chaile custom index ow add korte pari.
# element jei koyta custom index ow oi koytai hoite hobe. naile error ashbe.

custom_index = ['a','b','c','d','e']

#custom value uporer cell e ase

pd_series2 = pd.Series(data = data, index = custom_index)
pd_series2
```

```python
# eta onekta dictionary er moto kaj kore. dictionary er key er jaygay ekhane index diye value access korte hoy.

pd_series2['a']
```

### dictionary pass in Pandas Series

```python
# ami jodi series e ekta dictionary pass kore data hishebe tahole amake r alada kore index pass korte hobe na. karon dictionary er key kei index hishbe use korbe.

movie_dictionary = {
    'harry potter': 'fantasy',
    'avengers': 'action',
    'the nun': 'horror',
    'the fault in our stars': 'romance',
    'game of thrones': 'fantasy'
}

movie_series = pd.Series(data = movie_dictionary)
movie_series
```
