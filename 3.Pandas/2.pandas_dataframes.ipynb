## Dataframe selection using loc and iloc

```python
df3 = pd.DataFrame(
    data = np.random.randint(1, 100, size=(6, 5)),
    index = ['row1', 'row2', 'row3', 'row4', 'row5', 'row6'],
    columns = ['col1', 'col2', 'col3', 'col4', 'col5']
)

df3
```

### loc attribute

```python
# loc means location

# row select
df3.loc['row2'] # etaw ekta dataframe series return korbe.
```

```python
# column select
df3['col2'] # etaw ekta dataframe series return korbe.
```

```python
# row and column select
df3.loc['row2', 'col2'] 
```

```python
df3
```

```python
# ekhon ami 55,50, 97, 82 dekhte chai. mane row2 and row3 er col2 and col3 dekhte chai.

df3.loc[['row2', 'row3']]
```

```python

df3.loc[['row2', 'row3'], ['col2', 'col3']]
```

### iloc attribute

```python
# iloc means integer location.

df3
```

```python
# iloc diye jokhon selection korbo tokhon row and column ke integer value diye select korte hobe. also output ew row and column er nam integer value te show korbe.

# single element select:
df3.iloc[1, 1] # row2 and col2 er value show kor
```

```python
# multiple row and column select:

df3.iloc[[1,2]]
```

```python
df3.iloc[[1,2], [1,2]]
```

## Conditional Selection

```python
data= np.random.randn(8, 5)

df = pd.DataFrame(
    data=data,
    index=['row1', 'row2', 'row3', 'row4', 'row5', 'row6', 'row7', 'row8'],
    columns=['col1', 'col2', 'col3', 'col4', 'col5']
)

df
```

```python
# jei value gula 0 er theke boro tader dekhte chai

df[df>0]
```

```python
# column 3 er jei value gula 0 er theke boro tader dekhte chai
df[df['col3']>0]
```

```python
# multiple conditional selection:

# pandas e multi conditional selection e ektu careful thakte hobe. python or cpp te amra contion apply kori logical and, logical or diye. but pandas e kintu bitwise and(&), bitwise or(|) use korte hobe.

# like ami column 3 and column 5 both column er jei value gula positive shudhu tader dekhabe.

df[(df['col3']>0) & (df['col5']>0)] # biwsie operator use korte must be bracket dite hobe condition e.
```

```python
# ami uporer code e  bitwise and(&) and use korsi. ami jodi logical and use kortam tahole error show korto.

df[(df['col3']>0) and (df['col5']>0)]
```

```python
# uporer error er reason holo logical and, logical or just boolean value niye kaj kore. but ekhane amra duita column er upor condition apply korsi. and ei column gula holo ekta series.

# so ekhane compare hocche: pandas series vs pandas series. tai amader bitwise operators use korte hobe.

type(df['col3']>0)
```
