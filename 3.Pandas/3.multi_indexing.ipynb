```python
import pandas as pd
import numpy as np
```

```python
groups = ['group1'] * 3 + ['group2'] * 3
groups
```

```python
inside_groups = ['row1', 'row2', 'row3'] * 2
inside_groups
```

```python
# zip() function duita list ke combine kore tuple banay.

zip(groups, inside_groups)
```

```python
# only zip() function use korle amake just ekta object er memory location dekhabe. but ami jodi list() function er moddhe zip() function ke use kori tahole amake tuple er list dekhabe.

tup = list(zip(groups, inside_groups))
tup
```

```python
# ekhn ami ei list ke multiindex e convert korbo.

mul_index = pd.MultiIndex.from_tuples(tuples= tup)
mul_index
```

```python
# ekhon ei multiindex ke ami ekta dataframe e convert korle shundor ekta table dekhbo.

df = pd.DataFrame(
    data = np.random.randn(6, 2),
    index = mul_index,
    columns = ['col1', 'col2']
)

df
```

```python
# amra etokhon dhore ei table or dataset ta banailam. real life e eirokom dataset amra khub komi pabo. but oi dui ekbar jodi pai tokhon kivabe indexing korbo ta shekhar jonnei ei table ta eto jhamela kore banailam.

# ekhon group1 er row1 er value gula dekhbo.

df.loc['group1', 'row1']
```

```python
# ami group1 er row1 and group2 er row1 er value gula dekhte chai. 
df.loc[('group1', 'row1'), ('group2', 'row1')] # evabe access kora jabe na. error ashbe. solution hishbe ami group column er nam and row column er nam define kore dibo. then nam dhore indexing korbo.
```

```python
df.index.names = ['groups', 'rows']
df
```

```python
# ekhon cross section method (xs) use kore ami group wise alada alada indexing korte parbo.

# level er nam dhore xs method e indexing kora jay.

df.xs('row1', level='rows')
```
