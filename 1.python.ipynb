### Basic: Python-এর  Object, Reference এবং Identity: কমপ্লিট স্টাডি নোট

### ১. Object, Memory Location এবং Reference কী?
Python অন্যান্য ল্যাঙ্গুয়েজ (যেমন C/C++) থেকে একটু আলাদাভাবে কাজ করে। এখানে variable নিজে কোনো ডেটা স্টোর করে না, বরং ডেটাকে নির্দেশ বা point করে।
* **Memory Location:** কম্পিউটারের RAM-এর সেই নির্দিষ্ট ঠিকানা, যেখানে কোনো ডেটা ফিজিক্যালি সেভ করা থাকে।
* **Object:** Memory Location-এ সেভ করা আসল ডেটাটিই হলো Object। একে(memory location ke) একটি "বাক্স" হিসেবে কল্পনা করতে পারো যার ভেতরে ডেটা রাখা আছে।
* **Reference (Variable):** Variable নিজে কোনো বাক্স নয়; এটি হলো সেই বাক্সের গায়ে লাগানো একটি "নাম" বা "স্টিকার"। এটি শুধু বলে দেয় ডেটাটা মেমোরির কোথায় আছে।

**উদাহরণ:**

```python
arr = [1, 2, 3]

```
এখানে `[1, 2, 3]` হলো একটি **List Object** যা মেমোরিতে তৈরি হয়েছে, আর `arr` হলো একটি **Reference** যা ওই লিস্টকে point করে আছে।

### ২. "Everything is an Object" কনসেপ্ট
Python-এ তুমি যা-ই তৈরি করো না কেন, মেমোরিতে সবকিছুকেই Object হিসেবে ট্রিট করা হয়।

* সংখ্যা (`5`, `10.5`) → Integer/Float Object
* টেক্সট (`"Hello"`) → String Object
* লিস্ট/অ্যারে (`[1, 2]`) → List Object
* এমনকি ফাংশন (`def my_func():`) → Function Object

### ৩. `==` বনাম `is` Operator
এই দুটি operator সম্পূর্ণ ভিন্ন উদ্দেশ্যে কাজ করে:
* **`==` (Value Equality):** এটি শুধু চেক করে দুটি Object-এর **ভেতরের ভ্যালু বা ডেটা** একই কি না।
* **`is` (Object Identity):** এটি চেক করে দুটি variable মেমোরিতে **একই Object-কে** reference করছে কি না।

### ৪. `id()` ফাংশন দিয়ে Memory Location চেক করা
Python-এ প্রতিটি Object তৈরি হওয়ার সময় তার একটি unique ID বা মেমোরি ঠিকানা তৈরি হয়। `id()` ফাংশন দিয়ে আমরা সেই ঠিকানা দেখতে পারি।
মূলত, `is` operator ব্যাকএন্ডে এই `id()` চেক করেই True বা False রিটার্ন করে।

**উদাহরণ:**
```python
list1 = [1, 2]
list2 = [1, 2]
print(id(list1))  # Output: 140321456 (ধরে নিচ্ছি)
print(id(list2))  # Output: 140321900 (ভিন্ন ঠিকানা)

print(list1 == list2)  # True (কারণ ভেতরের ভ্যালু একই)
print(list1 is list2)  # False (কারণ মেমোরিতে তারা দুটি আলাদা Object)

```

### ৫. Exception: Integer Caching (বা Interning)
সাধারণ নিয়মে আলাদা variable-এ একই ভ্যালু রাখলে তাদের আলাদা Object হওয়ার কথা। কিন্তু ছোট সংখ্যার ক্ষেত্রে Python অন্যভাবে কাজ করে।

```python
a = 5
b = 5
print(a is b)  # Output: True (Exception!)
```
**কেন True হলো?**
এর কারণ হলো **Integer Caching**।
* প্রোগ্রামিংয়ে ছোট সংখ্যা অনেক বেশি ব্যবহার হয়। তাই মেমোরি এবং পারফরম্যান্স বাঁচানোর জন্য Python **`-5` থেকে `256**` পর্যন্ত সংখ্যাগুলোকে প্রোগ্রাম রান করার আগেই মেমোরিতে তৈরি করে রাখে (Cache করে রাখে)।
* তুমি যখন `a = 5` এবং `b = 5` লেখো, Python নতুন কোনো Object তৈরি করে না। সে মেমোরিতে আগে থেকেই থাকা ওই একই `5` Object-এর reference `a` এবং `b` উভয়কেই দিয়ে দেয়।
* যেহেতু দুটি variable-ই মেমোরির একই ঠিকানাকে point করে, তাই তাদের `id()` এক হয় এবং `is` operator True রিটার্ন করে।

**প্রমাণ (বড় সংখ্যার ক্ষেত্রে Caching কাজ করে না):**
```python
x = 1000
y = 1000
print(x is y)  # Output: False
```
এখানে `1000` সংখ্যাটি `-5` থেকে `256` রেঞ্জের বাইরে। তাই Python `x` এবং `y` এর জন্য মেমোরিতে দুটি সম্পূর্ণ আলাদা `1000` Object তৈরি করে, ফলে তাদের memory location আলাদা হয়।

**💡 এককথায় সামারি (Quick Recap):**

1. **Object:** মেমোরিতে রাখা আসল ডেটা।
2. **Reference:** সেই ডেটার মেমোরি ঠিকানা নির্দেশকারী নাম (variable)।
3. **`==` :** ডেটার ভ্যালু চেক করে।
4. **`is` :** ডেটার মেমোরি লোকেশন বা ID চেক করে।
5. **Integer Caching:** `-5` থেকে `256` পর্যন্ত সংখ্যাগুলো মেমোরিতে আগে থেকেই তৈরি থাকে, তাই এদের ক্ষেত্রে `is` সবসময় True হয়।

````python
# is operator er nam holo identity operator. 

# eta check kore duita variable er memory location same kina. duita variable er value same kina eta check kore na. value check kore equal (==) operator.

# `==` operator দুটি object-এর value সমান কি না check করে, আর `is` operator দুটি variable একই object-কে reference করছে কি না check করে। সহজভাবে `is`-কে একই memory location-এর object কি না check করা হিসেবে বোঝা যায়, তবে technically এটি object identity check করে। যেমন, a = 5 এবং b = 5 হলে Python ছোট integer cache করার কারণে a ও b একই `5` object-কে reference করতে পারে, তাই `a is b` True হয়। Value compare করার জন্য `==` এবং object identity compare করার জন্য `is` ব্যবহার করা উচিত।

a = 5
b = 5
````

````python
a == b 
````

````python
a is b # ekhane to false howar kotha. but eta exception.
````
